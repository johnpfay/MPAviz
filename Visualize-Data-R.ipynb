```r
#Import libraries
library(sf)
library(tidyr)
library(ggplot2)
```

```r
#Import data
fname_output <- '../scratch/Hexagons_from_R.shp'
fcHex <- st_read(fname_output,quiet=TRUE)
```

```r
#Extract dataframe
dfHex <- as.data.frame(fcHex)
```

```r
#Tidy the data for plotting 
dfHex2 <- dfHex %>% 
 select(c('NeedTotal','OppsTotal')) %>%
 pivot_longer(c('NeedTotal','OppsTotal'),
                            names_to='Type',
                            values_to='Count')
```

```r
#Histograms
ggplot(dfHex2, aes(x=Count,fill=Type)) + 
 geom_histogram(binwidth=1, alpha=0.5, position="dodge")
```

```r
#Density plots
ggplot(dfHex2, aes(x=Count, colour=Type)) + geom_density()
```
